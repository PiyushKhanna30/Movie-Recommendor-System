To find the similarity among different movies depend on type based.

```python
import pandas as pd
import numpy as np
```

```python
#RATING BY USER ACCORDING TO MOVIE ID
#columns naming for rating df
rating_cols=['user_id','movie_id','rating']
ratings =pd.read_csv("data/u.data",sep="\t",usecols=range(3),names=rating_cols,encoding="ISO-8859-1")
ratings.head()
```

```python
#MOVIE TITLE AND ID 
movie_cols=['movie_id','movie_name']
movies=pd.read_csv("data/u.item",sep="|",usecols=range(2),names=movie_cols,encoding="ISO-8859-1")
movies.head()
```

```python
#MERGING BOTH DATAFRAMES TO ONE 
ratings = pd.merge(movies, ratings)
ratings.head()
```

```python
#NOW ROTATING TABLE WITH USER_ID ON LEFT SIDE AND MOVIE NAME ON TOP  BASED ON RATING VALUES
movie_ratings=ratings.pivot_table(index=['user_id'],columns=['movie_name'],values='rating')
movie_ratings.head()
```

```python
#CORRELATION BETWEEN DIFFERENT MOVIES USING CORR FXN OF PANDAS
corr_matrix=movie_ratings.corr()
corr_matrix.head()
```

```python
#CORRELATION BETWEEN DIFFERENT MOVIES USING CORR FXN OF PANDAS BUT BETWEEN THEM WHERE ATLEAST 100 USERS RATED MOVIE
corr_matrix=movie_ratings.corr(min_periods=100)
corr_matrix.head()
```

```python
#NOW CREATING RECOMMENDATIONS FOR USER 1
user=1
user_ratings=movie_ratings.loc[user].dropna()
user_ratings
```

```python
#FINDING SIMILARITY OF SERIES WITH USER RATINGS 
#NOTE WHEN RATING 1 THEN USER DONT LIKE MOVIE SO WE DONT SHOW THAT
similar=pd.Series()
#appending all user movies atleast one so that we can drop even if movie not corr in corr_matrix to itself
similar=similar.append(user_ratings)
for i in range(0,len(user_ratings)):
    print("Adding similar movies depending on {}.".format(user_ratings.index[i]))
    #here we get list of movies and their ratings related to our movie
    sim=corr_matrix[user_ratings.index[i]].dropna()
    #multiplying our movie rating with the correlated movie
    sim=sim.map(lambda x: x * user_ratings[i])
    #appending this to series
    similar=similar.append(sim)
```

```python
print("Similar movies...")
similar.sort_values(inplace=True,ascending=False)
print(similar.head(10))
```

```python
#SINCE MULLTIPLE NAMES APPEAR IN SIMILAR SO WE NEED TO VALIDATE THAT ALSO USER MIGHT HAVE WATCH MOVIE THAT HE RATED SO WE NEED TO EXCLUDE IT
similar=similar.groupby(similar.index).sum()
similar.head(10)
```

```python
#SORTING THE SIMILAR BY THE SIMILAR IN DESC ORDER OF SIMILARITY
similar.sort_values(inplace=True,ascending=False)
similar.head(10)
```

```python
#NOW REMOVING MOVIES THAT USER HAS ALREADY WATCHED
final_movies=similar.drop(user_ratings.index)
final_movies
```
