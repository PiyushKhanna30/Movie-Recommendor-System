To find the similarity among different movies based on the user ratings.

```python
import pandas as pd
import numpy as np
```

```python
#RATING BY USER ACCORDING TO MOVIE ID
#columns naming for rating df
rating_cols=['user_id','movie_id','rating']
ratings =pd.read_csv("data/u.data",sep="\t",usecols=range(3),names=rating_cols,encoding="ISO-8859-1")
ratings.head()
```

```python
#MOVIE TITLE AND ID 
movie_cols=['movie_id','movie_name']
movies=pd.read_csv("data/u.item",sep="|",usecols=range(2),names=movie_cols,encoding="ISO-8859-1")
movies.head()
```

```python
#MERGING BOTH DATAFRAMES TO ONE 
ratings = pd.merge(movies, ratings)
ratings.head()
```

```python
#NOW ROTATING TABLE WITH USER_ID ON LEFT SIDE AND MOVIE NAME ON TOP  BASED ON RATING VALUES
movie_ratings=ratings.pivot_table(index=['user_id'],columns=['movie_name'],values='rating')
movie_ratings.head()
```

```python
#get ratings of star wars
starwars_ratings = movie_ratings['Star Wars (1977)']
starwars_ratings.head()
```

```python
#Finding movies similar to the star wars and putting in new dataframe
similar=movie_ratings.corrwith(starwars_ratings)
similar=similar.dropna()
similar.head()
```

```python
#SORTING SIMILAR IN ORDER SUCH THAT 1.00 SHOWS MOST RELATED AND -1.00 NOT RELATED
similar.sort_values(ascending=False)
```

However this is not perfect as there may be case where due a single user, he may like scify and then action but it is not tightly related. so we need to set that if certain number of user say>100.

```python
#HERE WE GROUP MOVIES BY TITLE THEN IN SIZE NO. OF PEOPLE WHO RATED AND MEAN IS THE AVERAGE OF RATINGS OUT OF 5
movie_ratings=ratings.groupby("movie_name").agg({'rating':[np.size,np.mean]})
movie_ratings.head(10)
```

```python
#NOW PREDICTING SIMILAR IF RELATED AS PER RATING BY SAY 100 PEOPLE
popular=movie_ratings['rating']['size']>=100
# popular has true where =>100 and False where <100
movie_ratings[popular].head()
```

```python
movie_ratings[popular].sort_values([('rating','mean')],ascending=False).head()
```

```python
#NOW COMBINING IT WITH THE COR DATAFRAME i.e. similar
df=movie_ratings[popular].join(pd.DataFrame(similar,columns=["similarity"]))
df.head(10)
```

```python
df.sort_values(['similarity'],ascending=False).head(10)
```
